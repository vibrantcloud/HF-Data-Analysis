```python
df2 = (
    df.groupby(["billing_id", "item_id"])["item_id"]
    .sum()
    .unstack()
    .reset_index()
    .fillna(0)
    .set_index("billing_id")
)
```

```python
df2.head(5)
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

basket_sets = df2.applymap(encode_units)
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.07, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head()
```
